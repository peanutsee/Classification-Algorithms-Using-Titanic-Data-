Allocate an arbituary cabin class to each passengers based on the fare they paid. Do so by using range method. 

0 - 20 == Class D
21 - 80 == Class C
81 - 150 == Class B
151 - xxx ==  Class A 

```python
df.drop(labels= 'Cabin', axis= 1, inplace= True)
```

```python
df['Fare']
```

```python
def converter(fare):
    if fare <= 20:
        return 'D'
    elif fare <= 80:
        return 'C'
    elif fare <= 150:
        return 'B'
    else:
        return 'A'
```

```python
passenger = df['Fare'].apply(converter)
```

```python
passenger_df = pd.DataFrame(passenger)
```

```python
passenger_df.rename(axis= 'columns', mapper= {'Fare':'Cabin'}, inplace= True)
passenger_df
```

```python
df = pd.concat([df, passenger_df], axis= 1)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('Survived', axis = 1)
y = df['Survived']
```

```python
X_unscaled = X
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X_unscaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size= 0.33, random_state= 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### Logistic Regression

```python
log = LogisticRegression()

print(log.fit(X_train, y_train))

pred = log.predict(X_test)

cm = confusion_matrix(y_test, pred)
```

```python
cm
```

```python
cm[0]
```

```python
TP = cm[0][0]
FP = cm[0][1]
TN = cm[1][1]
FN = cm[1][0]
```

```python
print(TP, FP, TN, FN)
```

```python
cm_matrix = np.array([[TP, FP], [FN, TN]])
```

```python
ax = sns.heatmap(cm_matrix, annot= True, cbar= False, xticklabels= False, yticklabels= False)
ax.set_ylim([0,2])
ax.invert_yaxis()
```

#### SVM

```python
from sklearn.svm import SVC
```
